```python
# 2014 
addToUrls = [
'http://dx.doi.org/10.1007/978-3-319-10470-6_58',
'http://dx.doi.org/10.1007/978-3-319-10470-6_39',
'http://dx.doi.org/10.1007/978-3-319-10443-0_39',
# 2015
'https://doi.org/10.1007/978-3-319-24553-9_5',
'https://doi.org/10.1007/978-3-319-24553-9_63',
'https://doi.org/10.1007/978-3-319-24553-9_69',
'https://doi.org/10.1007/978-3-319-24553-9_70',
'https://doi.org/10.1007/978-3-319-24553-9_74',
'https://doi.org/10.1007/978-3-319-24553-9_83',
'https://doi.org/10.1007/978-3-319-24553-9_87',
'https://doi.org/10.1007/978-3-319-24571-3_3',
'https://doi.org/10.1007/978-3-319-24571-3_9',
'https://doi.org/10.1007/978-3-319-24574-4_1',
'https://doi.org/10.1007/978-3-319-24574-4_27',
'https://doi.org/10.1007/978-3-319-24574-4_78',
'https://doi.org/10.1007/978-3-319-24574-4_80',
'https://doi.org/10.1007/978-3-319-24574-4_81',
'https://doi.org/10.1007/978-3-319-24553-9_72',
'https://doi.org/10.1007/978-3-319-24571-3_8',
'https://doi.org/10.1007/978-3-319-24574-4_28',
'https://doi.org/10.1007/978-3-319-24574-4_43',
'https://doi.org/10.1007/978-3-319-24553-9_62']
```

```python
urls.extend(addToUrls)
len(urls)
```

```python
# search result page
allTitles = []
allAuthors = []
allFullURLs = []
# abstract page
allEmails = [] 
allYears = []
allUses = []
allAbstracts = []
allComments = [] # combine with year into shortDetails
# extra one
allPdfURLs = []
#added at pandas level
# source = arvix
# key = uuid

def getName(authors,idx):
    return authors[idx].text.replace(u'\xa0', u' ').encode('ascii','ignore')

searchWords = ["deep", "encoder" , "learning", "convolutional", "neural" , "cnn", 'recurrent' , 'adversarial' , "rnn"]

for url in urls:
    try:
        print url
        # fetch url
        response1 = urllib2.urlopen(url)
        # convert to bs
        soup1 = bs.BeautifulSoup(response1,"html")

        # 1 get title
        titles = soup1.findAll("h1", { "class" : "ChapterTitle" })
        title = titles[0].text.replace(u'\xa0', u' ').encode('ascii','ignore')
        print title
        time.sleep(1)
        
        # check if we want it
        if any ( word in title.lower() for word in searchWords  ):
            
            # 1 get title
            allTitles.append( title  )
            # 2 get authors
            authors = soup1.findAll("span", { "class" : "authors__name" })
            authorString = getName(authors,0)
            for j in range(1,len(authors)):
                authorString += ', ' + getName(authors,j)
            allAuthors.append(authorString+'.')
            # 3 get all fullURLs
            allFullURLs.append ( url )
            # 4 get emails 
            emails = soup1.findAll("a", {"class" : "gtm-email-author"})
            if len(emails) == 0:
                allEmails.append([])
            else:
                allEmails.append([emails[0]['title']])
            # 5 get year
            year = soup1.findAll("a", { "class" : "gtm-book-link" })
            yr = year[0].text[year[0].text.strip().find('201')
                               :
                               year[0].text.strip().find('201')+4] 
            allYears.append(int(yr))
            # 6 get use based on year
            allUses.append(1)  
            # 7 get abstracts
            abstracts = soup1.find_all("p" , {'class' : 'Para' })
            if len(abstracts) == 0:
                allAbstracts.append("")
            else:
                allAbstracts.append( abstracts[0].text.replace(u'\xa0', u' ').encode('ascii','ignore').rstrip() )  
            
            # 8 get comments
            allComments.append(year[0].text.replace(u'\xa0', u' ').encode('ascii','ignore'))

            time.sleep(1)
        
    except urllib2.HTTPError:
        print url + " no http"
        pass

print "done"
    
```

```python
lista = [allTitles,allAuthors,allFullURLs,allComments,allYears,allAbstracts,allEmails,allUses,allPdfURLs]
for lis in iter(lista):
    print len(lis)
```
